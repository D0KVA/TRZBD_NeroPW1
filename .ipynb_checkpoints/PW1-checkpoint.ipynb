<h4>Теперь необходимо удалить и заменить значения.</h4>

В данной ситуации я решил, что будет хорошим выходом:
1. Удалить значения не рейтинговых игр, чтобы не учитывать их в данном исследовании (или что это)
2. Заменить английские значения названия команд шахмат на русские

```python
df = df[df['rated'] != False] #Тут датафрейм меняется так, что из старого из столбца рейтинговые или нет удаляются все значения, которые содержат False
```

```python
df #Подтверждаю работу
```

```python
#df['winner'] = df['winner'].replace("black", "Чёрные") - Это не сработало, поэтому я решил обратиться за помощью
df.loc[df['winner'] == "black", 'winner'] = "Чёрные"
df.loc[df['winner'] == "white", 'winner'] = "Белые" 
#Тут проиходит замена всех английских значений на русские
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Для начала необходимо импортирвоать библиотеки, чтобы работать с графиками
import warnings #Это используется, чтобы убрать ошибки, которые могут направить на путь дурной
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
#plt.figure(figsize=(20,10)) - Можно сделать так, чтобы было более обширно и круто
sns.countplot(x=df['victory_status'], palette='rainbow')
```

Тут выводится соотношение количества и исходов. (Это выполняет первый пункт)

* А самым маленьким из признаков является ничья (draw), она произошла где-то чуть меньше 1000 раз.


<h3>Чтобы выполнить второй пункт необходимо использовать barplot, который используется для сравнения значений между различными категориями</h3>

```python
sns.barplot(x=df['victory_status'],y=df['turns'], palette='rainbow') 
plt.title("Соотношение исходов к ходам") #Название графика
"""
Тут происходит сравнение двух значений, чтобы понять сколько ходов было сделано в какой-либо игре/ситуации и происходит сравнение того, как 
ходы взаимодействуют с исходом 
"""
```

1. Наибольшее среднее значение находится в ничьей - это что-то в районе 90
2. Наибольший доверительный интервал также находится в столбце ничьи - он составляет где-то 10
3. В данной ситуации наименьшим средним значением является столбец, когда сдаются - он составляет где-то 55-58 ходов

```python
plt.figure(figsize=(30,20))
sns.boxplot(x='opening_name', y='turns', data=df[df['opening_name'].isin(['Nimzowitsch Defense: Kennedy Variation', "Queen's Pawn", 
'Pirc Defense', 'Dutch Defense', 'Blackmar-Diemer Gambit: Pietrowsky Defense', 'Reti Opening', 'Sicilian Defense: McDonnell Attack'])], palette='rainbow') 
plt.title("Усы, которые показывают соотношение дебютов к количесвту ходов")
plt.xlabel("Дебюты", fontsize=20)
plt.ylabel("Количество ходов", fontsize=20)
#Название графика
#Тут просматривается закономерность того сколько играются матчи, которые начинались с разных дебютов
```

* Тут наибольшая медиана находится у "Queen's Pawn". У данного дебюта медиана составляет где-то 63-65 ходов
* Получается, что наибольшее количество выбросов тут у "Sicilian Defense: McDonnell Attack" и "Queen's Pawn"
* А наименьшее количество выбросов у всех остальных

```python
data52 = df.drop(columns=['white_id', 'black_id', 'winner', 'victory_status', 'created_at', 'rated', 'moves', 'increment_code', 'last_move_at', 'opening_eco', 'opening_ply', 'opening_name'])
data52 
'''
Удаление лишкних столбцов, чтобы была более адекватная и рабочая тепловая карта
'''
```

```python
sns.heatmap(data52.corr(), annot=True, cmap='cool') #Тепловая карта
'''
annot - показывает точное соотношение в циферках
cmap - придаёт красивую раскраску
'''
```

Тут очень слабая корреляция между рейтингами и ходами (0.15 - наименьшая), а вот между рейтингами белых и чёрных (а именно 0,71).
Это означает, что рейтинги влияют на количество ходов немного, а вот корреляция между рейтингами больше, что означает, что они влияют друг на друга очень сильно

Также надо описать холодные цвета, но тут самый холодный цвет - это *бирюзовый*
